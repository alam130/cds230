## 1. Matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1, 2.5, 3, 4])
plt.ylabel('Values')
plt.show() 
```

```python
plt.plot([1,2.5,3,4],linestyle='dotted',color='green')
plt.ylabel('Values')
plt.show() 
```

```python
plt.plot(range(10))
plt.title('Center Title')
plt.title('Left Title', loc='left')
plt.title('Right Title', loc='right')
plt.xlabel('time (s)')
plt.ylabel('Position')
plt.show()
```

```python
plt.plot(range(4), label="Line A") 
plt.plot([1,2.5,3,4], color="red", label="Line B")
plt.legend()
plt.show()
```

```python
x  = [1,4.5,6,3]
y  = [4.5,6,3,2]
plt.plot(x,y, marker='o')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.show()
```

```python
x  = [1,4.5,6,3, 2]
y  = [4.5,6,3,2, 2.5]
plt.scatter(x,y,50,"y")
plt.show()
```

### Question 2

```python
def Interest(P, r, n, t):
    inner = 1 + r / n
    power = n * t
    B = P * inner ** power
    return B
```

```python
def Interest2(P, r, n, t):
    B = P * (1+r/n) ** (n*t)
    return B
```

```python
Interest(1000, 0.025, 1, 5)
```

```python
Interest2(1000, 0.025, 1, 5)
```
