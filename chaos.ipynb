## Chaotic difference equation example

```python
import matplotlib.pyplot as plt
```

```python
def chaos(r,xn):
    return r * xn * (1-xn)
```

```python
# variables
r=2
x0=0.1
```

```python
n = 100
```

### r=2

```python
x0=0.1
final_list = [x0]
for i in range(n):
    x0 = chaos(r,x0)
    final_list.append(x0)
```

```python
plt.plot(final_list)
```
